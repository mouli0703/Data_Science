```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('IMDB Dataset.csv')
data
```

```python
data = data.sample(5000)
data
```

```python
data['sentiment'].unique()
```

```python
data['sentiment'].replace({'negative':0,'positive':1},inplace=True)
```

```python
data.head(5)
```

```python
data.info()
```

```python
import re 
```

```python
def clean_html(text):
    clean = re.compile('<.*?>')
    return re.sub(clean,'',text)
    
```

```python
data['review']=data['review'].apply(clean_html)
data
```

```python
data['review'].iloc[5]
```

```python
def remove_spl(text):
    x=''
    for i in text:
        if i.isalnum():
            x=x+i
        else:
            x=x+' '
    return x
```

```python
data['review']=data['review'].apply(remove_spl)
data
```

```python
data['review'].iloc[5]
```

```python
def convert_low(text):
    return text.lower()
```

```python
data['review']=data['review'].apply(convert_low)
data
```

```python
data['review'].iloc[5]
```

```python
from nltk.corpus import stopwords
```

```python
a=stopwords.words('english')
```

```python
def remove_stop(text):
    x=[]
    for i in text.split():
        if i not in a:
            x.append(i)
        else:
            pass
    return x       
```

```python
data['review']=data['review'].apply(remove_stop)
data
```

```python
data['review'].iloc[5]
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stem_words(text):
    x=[]
    for i in text:
        x.append(ps.stem(i))
    return x   
```

```python
data['review']=data['review'].apply(stem_words)
data
```

```python
data['review'].iloc[5]
```

```python
def join(list_input):
    return ' '.join(list_input)
```

```python
data['review']=data['review'].apply(join)
data
```

```python
data['review'].iloc[5]
```

```python
X = data['review']
y = data['sentiment']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
cv.fit_transform(data['review'])
```

```python
X = cv.fit_transform(data['review']).toarray()
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
nb = GaussianNB()
```

```python
nb.fit(X_train,y_train)
```

```python
y_pred = nb.predict(X_test)
```
